# Limpeza

```python
# Primeiro ser√° feito uma copia para n√£o correr o risco de modificar os dados originais
df_twitter_archive_clean = df_twitter_archive.copy()
df_tweet_json_clean = df_tweet_json.copy()
df_image_predictions_clean = df_image_predictions.copy()
```

## Aruma√ß√£o

## Juntar as Tabelas Primeiro

- Primeiro vamos juntar todas as informa√ß√µes em uma s√≥ tebela e descatar o que n√£o √© passivo de alguma an√°lise
### Code

```python
# Usando merge para juntar todas as tabelas deixando a "df_twitter_archive_clean" como refer√™ncia
df_twitter_archive_clean = pd.merge(left = df_twitter_archive_clean, right=df_tweet_json_clean, left_on='tweet_id', right_on='id', how='left')
# Combinando novamente mas agora apenas as linhas que tem n√∫mero de id iguais
df_twitter_archive_clean = pd.merge(left = df_twitter_archive_clean, right=df_image_predictions_clean, left_on='tweet_id',right_on = 'tweet_id', how = 'left')
# Remo√ß√£o de uma columna duplicada
df_twitter_archive_clean.drop(['id'],axis =1,inplace=True)
```

## Test

```python
df_twitter_archive_clean.info()
```

## Nas colunas "doggo" "floofer" "pupper" e "puppo" deveriam ser apenas uma coluna "dog_type'
### Code

```python
# Vamos extrair na coluna "text" qualquer str que possa conter o tipo do cachorro
df_twitter_archive_clean['dog_type'] = df_twitter_archive_clean['text'].str.extract('(puppo|pupper|floofer|doggo)')
# Mudando o tipo de varivel para categ√≥rica
df_twitter_archive_clean['dog_type'] = df_twitter_archive_clean['dog_type'].astype('category')
# Deletando columunas que n√£o seram usadas
df_twitter_archive_clean.drop(['doggo','floofer','pupper','puppo'],axis = 1,inplace = True)
```

### Extraindo as ra√ßas dos cahorros do classificador e deletando o que n√£o √© util

```python
# criando uma lista vazia para as ra√ßas dos cachorros
dog_race = []
# criando uma fun√ß√£o para extrair as ra√ßas dos cahorros e adicionar na lista
def dog_race_prediction(df):
    if df['p1_dog']== True:
        dog_race.append(df['p1'])
    elif df['p2_dog'] == True:
        dog_race.append(df['p2'])
    elif df['p3_dog'] == True:
        dog_race.append(df['p3'])
    else:
        dog_race.append('NaN')
# aplicando a fun√ß√£o em todas as columnas
df_twitter_archive_clean.apply(dog_race_prediction, axis = 1)
# criando nova coluna com a lista das ra√ßas
df_twitter_archive_clean['dog_race'] = dog_race
# deletando o que n√£o vai ser usado
df_twitter_archive_clean.drop(['p1','p1_dog' ,'p1_conf','p2','p2_conf','p2_dog','p3','p3_conf','p3_dog'],axis=1,inplace = True)
```

## Qualidade
## Columa "name" estar imcompleta com "None"
- Bucar na coluna 'text' usando str.extract, os nomes ao inseridos no texto que est√£o depois das palasvras "named" e "name is", e substituir todos nomes errados por "NaN"
### Code

```python
# salvando em "nomes1" os nomes extraidos da columna "test" que segem o patrao regex ('name is\s(\w+)|named\s(\w+)')
# filtrando as linhas que tem "None" como observa√ß√£o na coluna "name"
names1 = df_twitter_archive_clean[df_twitter_archive_clean.name.str.contains('None') & \
                                  ((df_twitter_archive_clean.text.str.contains('named') |\
                                    df_twitter_archive_clean.text.str.contains('name is')))]\
                                    ['text'].str.extract(r'name is\s(\w+)|named\s(\w+)')
# criando uma colimna que os valores n√£o nulos dos 2 grupos do regex
names1.loc[names1[names1[1].notnull()].index,0] = names1[names1[1].notnull()][1]
# selecionando a coluna que est√£o todos os regex encontrados
names1 = names1[0]
# trocando os nomes com os repectivos indexes
df_twitter_archive_clean.loc[names1.index,'name'] = names1
```

```python
# o mesmo me repete aqui, mas com nomes que est√£o minusculos, ou seja, errados
names2 = df_twitter_archive_clean[df_twitter_archive_clean.name.str.islower() & \
                                  ((df_twitter_archive_clean.text.str.contains('named') | \
                                    df_twitter_archive_clean.text.str.contains('name is')))]\
                                    ['text'].str.extract(r'name is\s(\w+)|named\s(\w+)')
names2.loc[names2[names2[1].notnull()].index,0] = names2[names2[1].notnull()][1]
names2 = names2[0]
df_twitter_archive_clean.loc[names2.index,'name'] = names2
```

```python
# agora substituindo os que n√£o foram corigiodos por "NaN"
df_twitter_archive_clean.loc[df_twitter_archive_clean[df_twitter_archive_clean.name.str.islower()].index,'name'] = "NaN"
```

## Na Columna "rating_numerator", quando tem decimal,so estar pegando a parte decinal do n√∫mero

### Code

```python
# filtrando as linhas onde no texto da columna "text" contem o padr√£o regex '\d+\.\d*\/\d+' 
# salvando as posi√ß√µes
indexs = df_twitter_archive_clean[df_twitter_archive_clean.text.str.contains(r"(\d+\.\d*\/\d+)")].text.str.extract(r"(\d+\.\d*)").index
# salvando os valores completos
values = df_twitter_archive_clean[df_twitter_archive_clean.text.str.contains(r"(\d+\.\d*\/\d+)")].text.str.extract(r"(\d+\.\d*)").values
# substituindo pelo valor completo
df_twitter_archive_clean.loc[indexs].rating_numerator = values
```

# Visualisando e analisando os dados

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_twitter_archive_clean.info()
```

```python
c = df_twitter_archive_clean.corr()
sns.heatmap(c, 
            xticklabels=c.columns.values,
            yticklabels=c.columns.values,annot=True, linewidths=2)
plt.title('Mapa de Correla√ß√£o')
```

- Apenas 'retweet_count' e 'favorite_count' tem alguma correla√ß√£o
- Numero de imagem nao influencia nas notas

```python
df_twitter_archive_clean.plot(x = 'favorite_count' ,y ='retweet_count',kind = 'scatter')
plt.title('Rela√ß√£o entre favorite e retweet ')
plt.xlabel('Favorites')
plt.ylabel('Retweets')
```

- Confimando a correla√ß√£o entre 'retweet_count' e 'favorite_count', mostrando ser uma correla√ß√£o possitiva forte

```python
df_twitter_archive_clean.groupby(df_twitter_archive_clean['timestamp'].dt.year)\
                                [['retweet_count','favorite_count']].sum().plot(kind= 'bar')
plt.title("retweet_count e favorite_count por ano")
plt.ylabel("Quantidade")
plt.xlabel("Ano")
```

- Apesar do melhor ano ser 2017, n√£o teve muita diferen√ßa entre os anos de 2016 e 2017
- Aumente expressivo do ano 2015 para 2016
- Diminui√ß√£o de retweet em 2017 em rela√ß√£o a 2016, porem aumento de favorite

```python
df_twitter_archive_clean['dog_race'].value_counts(dropna= True)[1:6].sort_values().plot(kind = "barh", color = "blue",alpha = 0.7)
plt.title("Ra√ßas Mais Comuns")
plt.ylabel("Ra√ßa")
plt.xlabel("Frqu√™ncia")
```

- Ainda tem muitas ra√ßas que o classificador n√£o separou
- A ra√ßa mais comum √© a Golden Retriver

```python
df_twitter_archive_clean.groupby('dog_race')[['retweet_count','favorite_count']].sum().sort_values(by = 'retweet_count',ascending = False)[1:6].plot(kind = "bar")
plt.title("Ra√ßas mais Populares (Mais retweet e favoritadas)")
plt.ylabel("Quantidade")
plt.xlabel("Ra√ßa")
```

- Pode observar que os mais comuns tendem ser tamb√©m os mais poulares

```python
df_twitter_archive_clean['dog_type'].value_counts(dropna = True)[0:10].sort_values().plot(kind = "barh", color = "blue",alpha = 0.7)
plt.title("Tipo de cahorro mais comum")
plt.ylabel("Tipo")
plt.xlabel("Frequ√™ncia")
```

- Filhotes s√£o os mais comumns e provavelmente mais populares
